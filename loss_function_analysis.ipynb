```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(d_real, d_fake, x, z):
    z_opt = x_opt = 0
    loss = np.log(d_real) + np.log(1-d_fake)
    loss += (z - z_opt)**2
    loss += np.abs(x - x_opt)
    
    return loss
```

```python
d_real = 0.5
d_fake = 0.5
x = 0
z = 0
xvalues, yvalues = np.meshgrid(np.arange(0, 1, 0.02), np.arange(-2, 2, 0.02))
```

```python

loss_vect = [loss(d_real, 1-d_real, x, x) for d_real, x in zip(xvalues.flatten(), yvalues.flatten())]
loss_vect = np.reshape(loss_vect, xvalues.shape)
```

```python
plt.figure()
plt.contourf(xvalues, yvalues, loss_vect)
plt.colorbar()
plt.show()
```
